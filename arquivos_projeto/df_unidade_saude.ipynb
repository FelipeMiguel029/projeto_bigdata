```python
import pandas as pd
import unicodedata as ud

df = pd.read_csv("unidades_saude_noroeste_goiania.csv", sep=",")


df = df.applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)

df = df.drop_duplicates()
df = df.fillna("nao informado")



def remover_acentos(x):
    if isinstance(x, str):
        return "".join(
            c for c in ud.normalize("NFKD", x)
            if not ud.combining(c)
        )
    return x

df = df.applymap(remover_acentos)
print(df)


# load
df.to_csv("arquivo_tratado.csv", index=False)



```
